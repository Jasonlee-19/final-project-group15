### --Survey about the effect of family background on math achievement

```python
import streamlit as st
import pandas as pd
import matplotlib.pyplot as plt
plt.style.use('seaborn')
import math

df = pd.read_csv('Maths.csv')
```

```python
df.head()
```

### Q1: Effect of average parental education level on math achievement

```python
df.avedu.value_counts()
```

```python
df.groupby('avedu')['avegrade'].mean().plot.line()
```

### Q2: The impact of willing to take higher education on academic performance

```python
df.groupby('higher')['avegrade'].mean().plot.bar()
```

### *Alcohol consumption of student

```python
# Pie chart about workday alcohol consumption of student
df['Dalc'].value_counts(normalize=True).plot.pie(autopct='%.2f%%')
plt.title('Workday alcohol consumption of student\n--from [1:very low] to [5:very high]')
```

```python
# Pie chart about weekend alcohol consumption of student
df['Walc'].value_counts(normalize=True).plot.pie(autopct='%.2f%%')
plt.title('Weekend alcohol consumption of student\n--from [1:very low] to [5:very high]')
```
